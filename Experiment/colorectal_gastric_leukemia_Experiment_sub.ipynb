```python
def create_combined_df(cancer_df, normal_df, random_state):
    cancer_count = len(cancer_df)
    normal_sample = normal_df.sample(n=cancer_count, random_state=random_state, replace=False)

    combined_df = pd.concat([cancer_df, normal_sample], ignore_index=True)
    return combined_df
```

```python
df_colorectal_cancer = create_combined_df(df_colorectal, df_normal, 4)
df_gastric_cancer = create_combined_df(df_gastric, df_normal, 5)
df_leukemia_cancer = create_combined_df(df_leukemia, df_normal, 6)
```

## Sub-Type Classification using Logistic Regression

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix

def kfold_with_pca(X, y, n_components, n_splits=5):
    
    # Get the unique class labels
    unique_classes = np.unique(y)

    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    pca = PCA(n_components=n_components)
    model = RandomForestClassifier(n_estimators=100, random_state=42)

    overall_accuracies = []
    per_class_accuracies = []
    per_class_recalls = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Apply PCA
        X_train_pca = pca.fit_transform(X_train)
        X_test_pca = pca.transform(X_test)

        # Train the model
        model.fit(X_train_pca, y_train)

        # Predict on the test set
        y_pred = model.predict(X_test_pca)

        # Calculate overall accuracy
        overall_accuracies.append(accuracy_score(y_test, y_pred))

        # Calculate per-class metrics using complete class labels
        cm = confusion_matrix(y_test, y_pred, labels=unique_classes)
        class_accuracies = cm.diagonal() / cm.sum(axis=1, where=cm.sum(axis=1) > 0)  # Per-class accuracy
        per_class_accuracies.append(class_accuracies)

        class_recalls = recall_score(y_test, y_pred, average=None, labels=unique_classes, zero_division=0)
        per_class_recalls.append(class_recalls)

    # Convert metrics to NumPy arrays for easier aggregation
    per_class_accuracies = np.array(per_class_accuracies)
    per_class_recalls = np.array(per_class_recalls)

    # Calculate mean metrics
    mean_overall_accuracy = np.mean(overall_accuracies)
    mean_per_class_accuracy = np.nanmean(per_class_accuracies, axis=0)  # Handle NaN for missing classes
    mean_per_class_recall = np.mean(per_class_recalls, axis=0)

    # Final model training on the full dataset
    X_pca_full = pca.fit_transform(X)
    model.fit(X_pca_full, y)

    # Print results
    print(f"5-Fold CV Results with {n_components} PCA components:")
    print(f"Mean Overall Accuracy: {mean_overall_accuracy * 100:.2f}%")
    for i, (acc, rec) in enumerate(zip(mean_per_class_accuracy, mean_per_class_recall)):
        print(f"Type {unique_classes[i]} - Accuracy: {acc * 100:.2f}%, Recall: {rec * 100:.2f}%")

    # Return the final trained model and metrics
    return {
        "model": model,
        "pca": pca,
        "Mean Overall Accuracy": mean_overall_accuracy,
        "Mean Per-Class Accuracy": dict(zip(unique_classes, mean_per_class_accuracy)),
        "Mean Per-Class Recall": dict(zip(unique_classes, mean_per_class_recall)),
    }
```

```python
# Example: Test with different PCA components
for n in [10, 20, 50, 100, 256]:
    results = kfold_with_pca(Xs_c, ys_c, n_components=n)
    print(f"\nFinal Model Trained with {n} PCA Components:")
    print(results["model"])
    print()
```

```python
# Example: Test with different PCA components
for n in [2, 4, 8, 16, 32]:
    results = kfold_with_pca(Xs_g, ys_g, n_components=n)
    print(f"\nFinal Model Trained with {n} PCA Components:")
    print(results["model"])
    print()
```
