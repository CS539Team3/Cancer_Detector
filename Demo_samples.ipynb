```python
import pandas as pd

# Load the data
demo_data = pd.read_csv("Test_data/test_data.csv")

# Define the indices to select
index = [6, 48, 111]

# Select rows with the specified indices
selected_rows = demo_data.loc[index]

# Save each selected row as a separate CSV file
for idx, sample in selected_rows.iterrows():
    sample_df = sample.to_frame().T  # Convert Series to DataFrame
    file_name = f"selected_sample_{idx}.csv"  # Use the index in the file name
    sample_df.to_csv(file_name, index=False)  # Save the row as a separate CSV file
    print(f"Saved sample {idx} to '{file_name}'.")
```
